Title of Project : MOVIE RECOMMENDER SYSTEM

Objective : IT PREDICTS MOVIES TO USERS ACCORDING TO THEIR CHOICES

Data Source : GITHUB


```python
import pandas as pd
movieDf = pd.read_csv("https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Movies%20Recommendation.csv")
movieDf.head()
movieDf.info()
movieDf.columns
movieDf.describe()
movieDf.shape
features=movieDf[['Movie_Title', 'Movie_Genre','Movie_Keywords','Movie_Cast','Movie_Director']].fillna(' ')
features.shape
features.info()
x= features['Movie_Title']+" "+features['Movie_Genre'] +features['Movie_Keywords'] +" "+features['Movie_Cast'] +" "+features['Movie_Director']
x
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
x= tfidf.fit_transform(x)
x.shape
print(x)
from sklearn.metrics.pairwise import cosine_similarity
score =cosine_similarity(x)
score
score.shape
Movie_name=input("Enter Movie NAme:")
title_list=movieDf['Movie_Title'].tolist()
import difflib
recommendation = difflib.get_close_matches(Movie_name,title_list)
print(recommendation)
closeMatch=recommendation[0]
print(closeMatch)
index_of_closeMatch=movieDf[movieDf.Movie_Title==closeMatch]['Movie_ID'].values[0]
print(index_of_closeMatch)
recomend_score=list(enumerate(score[index_of_closeMatch]))
print(recomend_score)
len(recomend_score)
Sorted_similarly_movie=sorted(recomend_score,key=lambda x:x[1],reverse=True)
print(Sorted_similarly_movie)
print('You should Try Following 5 Movies')
i=1
for movies in Sorted_similarly_movie:
  index=movies[0]
  title_index = movieDf[movieDf.index==index]['Movie_Title'].values[0]
  if(i<6):
    print(i,' : ',title_index)
    i+=1

```
